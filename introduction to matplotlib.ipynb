##matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax =plt.subplots(figsize=(10, 10)) #(width, heigh)

# 3. plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple Plot', 
       xlabel='x-axis',
       ylabel='y-axis')
#5. Save & show (you save the whole figure)
fig.savefig('sample-plot.png')
```

```python
#Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter":8,
                     "Cashew butter":12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)")
```

```python
Fruits = {"Banana": 8,
          "Melon":10,
          "Grapes":20}
fig, ax = plt.subplots(figsize=(5,5))
ax.bar(Fruits.keys(), Fruits.values())
ax.set(title="Price Fruits",)
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots


```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                      figsize=(10,5))
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000))
```

## Which one should you use? (pyplot vs matplotlib OO method)
When plotting something quickly, okay to use the pyplot method
when plotting something more advanced, use the OO method
heart_disease

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,5))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
#ax.set_xlim([45, 100])
```
